```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def omega(k_):
    return np.sqrt(k_**2 + 1)
```

```python
k = np.linspace(0, 4, 100)
plt.plot(k, omega(k))
```

```python
def v_ph(k):
    return np.sqrt(k**2 + 1)/k
def v_gr(k):
    return k/np.sqrt(k**2 + 1)
```

```python
k = np.linspace(0.1, 2, 100)
plt.plot(k, v_ph(k))
plt.plot(k, v_gr(k))
```
